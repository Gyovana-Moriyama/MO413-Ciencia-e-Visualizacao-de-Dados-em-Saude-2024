```python
import gseapy as gp
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
metabolites = pd.read_csv('../../data/processed/metabolites_from_pathways.csv', sep=';')
```

Grafo com das redes enriquecidas e os metab√≥litos associados a elas

```python
G = nx.Graph() 
```

```python
G.add_nodes_from(metabolites['Pathway'])
G.add_nodes_from(metabolites['Query'])
```

```python
edges = [(row['Pathway'], row['Query']) for index, row in metabolites.iterrows()]
```

```python
G.add_edges_from(edges)
```

```python
#Plot do grafo
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=300, node_color='skyblue', font_size=10, font_color='black')
plt.show()
```

```python
nx.write_gml(G, "../../data/processed/pathway_metabolites_network.gml")
```
