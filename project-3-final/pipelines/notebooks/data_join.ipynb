```python
import pandas as pd
```

```python
positive = pd.read_csv('./MTBLS6039_129482_compressed_files/m_MTBLS6039_LC-MS_positive_reverse-phase_metabolite_profiling_v2_maf.tsv', sep='\t')
negative = pd.read_csv('./MTBLS6039_129482_compressed_files/m_MTBLS6039_LC-MS_negative_reverse-phase_metabolite_profiling_v2_maf.tsv', sep='\t')
subjects = pd.read_csv('./MTBLS6039_129482_compressed_files/s_MTBLS6039.txt', sep='\t')
```

```python
subjects = subjects[subjects['Factor Value[Cohort]'].isin(['Prostate cancer', 'Healthy'])][['Sample Name', 'Factor Value[Cohort]']]
subjects['Label'] = subjects['Factor Value[Cohort]'].apply(lambda x: 0 if x == 'Healthy' else 1)
subjects = subjects.T
subjects.columns = subjects.loc['Sample Name', :]
subjects = subjects.loc['Label', :]
```

```python
pos_met = positive['metabolite_identification'].tolist()
neg_met = negative['metabolite_identification'].tolist()

in_common = set(pos_met).intersection(set(neg_met))
in_common
```

```python
df = pd.concat([positive, negative], axis=0)
df = pd.concat([df['metabolite_identification'], df.filter(subjects.index)], axis=1)
df = pd.concat([subjects, df.T], axis=1).T
df.iloc[0, -1] = 'Label'
df.index = df.iloc[:, -1]
df.index.name = None
df.drop('metabolite_identification', axis=1, inplace=True)
df
```

```python
df.to_csv('../../data/external/data_original.csv')
```
