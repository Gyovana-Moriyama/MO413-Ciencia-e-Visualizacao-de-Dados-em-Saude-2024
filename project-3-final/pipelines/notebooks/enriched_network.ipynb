```python
import gseapy as gp
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
metabolites_up = pd.read_csv('../../data/processed/metabolites_from_pathways_upregulated.csv', sep=';')
metabolites_down = pd.read_csv('../../data/processed/metabolites_from_pathways_downregulated.csv', sep=';')
```

Grafo com das redes enriquecidas e os metab√≥litos associados a elas

```python
G_up = nx.Graph() 
G_down = nx.Graph() 
```

```python
G_up.add_nodes_from(metabolites_up['Pathway'])
G_up.add_nodes_from(metabolites_up['Query'])
```

```python
G_down.add_nodes_from(metabolites_down['Pathway'])
G_down.add_nodes_from(metabolites_down['Query'])
```

```python
edges_up = [(row['Pathway'], row['Query']) for _, row in metabolites_up.iterrows()]
G_up.add_edges_from(edges_up)
```

```python
edges_down= [(row['Pathway'], row['Query']) for _, row in metabolites_down.iterrows()]
G_down.add_edges_from(edges_down)
```

```python
#Plot do grafo up regulated
pos = nx.spring_layout(G_up)
nx.draw(G_up, pos, with_labels=True, node_size=300, node_color='skyblue', font_size=10, font_color='black')
plt.show()
```

```python
#Plot do grafo down regulated
pos = nx.spring_layout(G_down)
nx.draw(G_down, pos, with_labels=True, node_size=300, node_color='skyblue', font_size=10, font_color='black')
plt.show()
```

```python
nx.write_gml(G_up, "../../data/processed/pathway_metabolites_network_up.gml")
nx.write_gml(G_down, "../../data/processed/pathway_metabolites_network_down.gml")
```
